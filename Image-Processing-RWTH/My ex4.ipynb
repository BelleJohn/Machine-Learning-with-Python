```python
##Task 4: Graph-cut Segmentation
##reference:https://docs.opencv.org/3.4/d8/d83/tutorial_py_grabcut.html

%matplotlib inline
import cv2
import matplotlib.image as mpimg
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pylab as pylab

pylab.rcParams['figure.figsize'] = (15.0, 15.0)

img_tooth = cv2.imread('tooth.jpg')
img_melanoma = cv2.imread('melanoma.jpg')
plt.subplot(121), plt.imshow(img_melanoma)
plt.show()

# YOUR CODE HERE
#cimg_tooth = img_tooth
#cimg_melanoma = img_melanoma 

mask = np.zeros(img_tooth.shape[:2],np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (60,60,330,290)
cv2.grabCut(img_tooth,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
cimg_tooth = img_tooth*mask2[:,:,np.newaxis]

mask = np.zeros(img_melanoma.shape[:2],np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (60,60,330,290)
cv2.grabCut(img_melanoma,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
cimg_melanoma = img_melanoma*mask2[:,:,np.newaxis]

# END OF YOUR CODE

# Convert back to uint8 values
cimg_tooth = np.uint8(cimg_tooth)
cimg_melanoma = np.uint8(cimg_melanoma)

plt.subplot(121), plt.imshow(img_tooth)
plt.title('Original image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cimg_tooth)
plt.title('Segmented image'), plt.xticks([]), plt.yticks([])
plt.show()

plt.subplot(121), plt.imshow(img_melanoma)
plt.title('Original image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cimg_melanoma)
plt.title('Segmented image'), plt.xticks([]), plt.yticks([])
plt.show()
```
